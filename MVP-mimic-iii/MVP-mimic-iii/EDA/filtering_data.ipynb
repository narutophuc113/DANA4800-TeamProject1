This notebook takes the data extracted from the MIMIC-III database and accomplishes the following tasks:

1. Filter out :

* Patients outside the age range of 18 - 90
* Patients that are not on mechanical ventillation
* Irrelevant columns

2. Rename columns

--------------------------------------------------------------------------------

Input: final_dataset.csv

Output : filtered_data.csv

```python
import pandas as pd
import numpy as np
pd.set_option('display.max_columns', None)
```

```python
extract = pd.read_csv('final_dataset.csv')
extract.head()
```

# Filtering

## Row-wise

```python
### AGE

# keep only patients in the age range of 18 - 90 years old
df = extract[(extract['age_hosp_in'] >= 18) & (extract['age_hosp_in'] <= 90)]

# round ages to integer
df['age_hosp_in'] = df['age_hosp_in'].astype(int)
```

```python
### MECHANICAL VENTILATION

# keep only mechanically ventilated patients
df = df[df['mechvent'] == 1]
```
